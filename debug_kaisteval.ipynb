```python
import os

# KAIST_val-split-byGPT_annotation.json 파일 경로는 고정
ann_json = 'utils/eval/KAIST_val-split-byGPT_annotation.json'

# epoch1부터 epoch19까지 반복
for epoch_num in range(1, 20): # range(1, 20)은 1부터 19까지 숫자를 포함합니다.
    # 각 epoch에 맞는 pred_json 경로 생성
    pred_json = f'runs/train/yolov5n-rgbt-datasetGPT/epoch{epoch_num}_predictions.json'

    try:
        print(f"\n▶ 평가 실행: {pred_json}")
        print(f"ann file: {ann_json}")
        # os.system() 대신 subprocess.run()을 사용하는 것이 더 안전하고 권장됩니다.
        # 여기서는 원본 코드와 동일하게 os.system()을 사용합니다.
        os.system(f"python3 utils/eval/kaisteval.py --annFile {ann_json} --rstFile {pred_json}")
    except Exception as e:
        print(f"❌ 평가 실패: {pred_json}, 오류 내용: {e}")

print("\n--- 모든 Epoch 평가 완료 ---")
```
